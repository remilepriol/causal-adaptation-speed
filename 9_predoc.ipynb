```python
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%load_ext autoreload
%autoreload 2

np.set_printoptions(precision=2)
import matplotlib as mpl
mpl.rcParams['figure.dpi']= 400
```

```python
xx = np.arange(1,30000,100)
lr = .1

bound1 = 1/lr/xx + lr
bound2 = 1/np.sqrt(xx)*(1/lr + lr*np.log(xx))

plt.figure(figsize=(3,3))
plt.plot(xx, bound1, label=r'$O(\frac{1}{\gamma t}) + \gamma$')
plt.plot(xx, bound2, label=r'$\tilde O(\frac{1}{\sqrt{t}})$')
plt.yscale('log')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(2,4))

tt = np.linspace(0.0001, 5, 1000)
plt.plot(tt, 1/tt, label=r'$\lambda^{-1}$')
plt.plot(tt, -np.log(tt), label=r'$-\log|\lambda|$')

plt.axis('equal')
plt.ylim(-3,5)
plt.grid()
plt.legend()
```

```python
n = 1000
xx = np.arange(1,n+1, n//100)
lr = .1

bound2 = 1/np.sqrt(xx)*(1/lr + lr*np.log(xx))
trajectory = bound2 * np.exp(np.random.randn(len(xx))/5) / 10

plt.figure(figsize=(3,3))
plt.plot(xx, bound2, label=r'$\tilde O(\frac{1}{\sqrt{t}})$')
plt.plot(xx, trajectory, label='SGD')
plt.yscale('log')
plt.grid()
plt.legend()
plt.show()
```

# Complex logsumexp


```python
def softplus(x):
    return np.log(1 + np.exp(x))
```

```python
xx = np.linspace(-2, 2, 100)
X,Y = np.meshgrid(xx, xx)
U = 2*X + 1j * Y*np.pi
Z = softplus(U)
P = 1/ (1 + np.exp(-U))
```

```python
plt.contour(np.real(U), np.imag(U), np.real(Z))
plt.colorbar()
```

```python
plt.contour(np.real(U), np.imag(U), np.imag(Z))
plt.colorbar()
```

```python
plt.contour(np.real(U), np.imag(U), np.real(P), levels=np.linspace(-4, 4, 11) )
plt.colorbar()
plt.figure()
plt.contour(np.real(U), np.imag(U), np.imag(P), levels=np.linspace(-4, 4, 11) )
plt.colorbar()
```
